## 1. tokenize each review

```python

comments = pd.read_csv('IA2.csv', delimiter=',',header=None)
```

```python
comments.describe()
```

```python
comments.head()
```

```python
tokenized_comments = []

for reviews in comments[:][1]:
    tokenized_comments.append(nltk.word_tokenize(reviews))
    
print(tokenized_comments)
    
```

## 2. Stemming

```python
# stemmer
stemmer = nltk.stem.PorterStemmer()
stemmed_words = []
```

```python
# loop through each array within greater list and stem
for x in tokenized_comments:
    for y in x:
        stemmed_words.append(stemmer.stem(y))
```

```python
print(stemmed_words)
```

```python
lemmatizer = nltk.stem.WordNetLemmatizer()
```

```python
lemmatized_words = []
```

```python
# loop through each array within greater list and stem
for x in tokenized_comments:
    for y in x:
        lemmatized_words.append(lemmatizer.lemmatize(y))
```

```python
print(lemmatized_words)
```

## 3. Stop Words

```python
# import dimension reduction library
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
filtered_sentence = [w for w in lemmatized_words if not w in stop_words] 
```

```python
filtered_sentence = [] 
```

```python
# loop through lemmatized words and remove stop words
for w in lemmatized_words: 
    if w not in stop_words: 
        filtered_sentence.append(w) 
```
